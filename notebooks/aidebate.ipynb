```python
temperature=0.7

model = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=temperature)
# model = ChatOpenAI(model_name="gpt-4", temperature=temperature)
```

```python
# gov_motion = "This house would ban the use of facial recognition technology"
# gov_args = """- It is a violation of privacy
# - It is a violation of human rights
# - It is a violation of human dignity
# - It is a violation of human autonomy
# - It is a violation of human freedom
# - It is a violation of human agency
# """
# opp_args = """- This will lead to more crime
# - This will lead to more violence
# - This will stop AI from being used for good
# - This wll stop the progress in AI
# - This will stop the progress in technology"""

# gov_motion = "This house believes that we should privatize the prison system"
# gov_args = """- saving money
# - better quality of life for prisoners
# - better quality of life for guards"""
# opp_args = """- wrong incentives for guards
# - wrong incentives for prisoners"""


gov_motion = "Parliament members should be appointed by a lottery among the general population"
gov_args = """- Democratic Representation
- Reducing Corruption
- Eliminating Campaign Costs"""
gov_mod = """Create very emotional, dramatic, arguments."""
opp_args = """- Lack of Expertise
- Lack of Accountability
- Lack of Transparency
- Potential for Apathy"""
opp_mod = """Create very calm, ironic arguments."""
```

```python
gov_history = ChatMessageHistory()
opp_history = ChatMessageHistory()
```

```python
gov_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "In parliament debates, you are the government. You are arguing for the motion: {gov_motion}. Your reasons are:\n{gov_args}\n\n{gov_mod}"),
] + gov_history.messages)

gov_chain = gov_prompt_template | model
gov_speech = gov_chain.invoke({
    "gov_motion": gov_motion,
    "gov_args": gov_args,
    "gov_mod": gov_mod,
})

gov_history.add_ai_message(gov_speech.content)
opp_history.add_user_message(gov_speech.content)

print("====================================")
print("Government statement 1")
print("====================================")
print()
print(gov_speech.content)
print()
```

```python
for i in range(3):
    opp_prompt_template = ChatPromptTemplate.from_messages([
        ("system", "In parliament debates, I am the government and you are the opposition. You are arguing against my motion: {gov_motion}. Your reasons are:\n{opp_args}\n\nYou are also arguing against my points in the previous discussion.\n\n{opp_mod}"),
    ] + opp_history.messages)

    opp_chain = opp_prompt_template | model
    opp_speech = opp_chain.invoke({
        "gov_motion": gov_motion,
        "opp_args": opp_args,
        "opp_mod": opp_mod,
    })

    gov_history.add_user_message(opp_speech.content)
    opp_history.add_ai_message(opp_speech.content)

    print("====================================")
    print(f"Opposition statement {i+1}")
    print("====================================")
    print()
    print(opp_speech.content)
    print()

    gov_prompt_template = ChatPromptTemplate.from_messages([
        ("system", "In parliament debates, I am the opposition and you are the government. You are arguing for the motion: {gov_motion}. Your reasons are:\n{gov_args}\n\nYou are also arguing against my points in the previous discussion.\n\n{gov_mod}"),
    ] + gov_history.messages)

    gov_chain = gov_prompt_template | model
    gov_speech = gov_chain.invoke({
        "gov_motion": gov_motion,
        "gov_args": gov_args,
        "gov_mod": gov_mod,
    })

    gov_history.add_ai_message(gov_speech.content)
    opp_history.add_user_message(gov_speech.content)

    print("====================================")
    print(f"Government statement {i+2}")
    print("====================================")
    print()
    print(gov_speech.content)
    print()
```
